```python
from turtle import color
from  mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from numpy import row_stack
import pandas as pd
# Normally the data is input using csv files, but the data was getting corrupted
headers=['Open Time','High','Low']
df=pd.read_csv('../input/bitcoin/Bit.csv')
day=df["day"]
high=df["High"]
low=df["Low"]
x=[]
y=[]
z=[]
x=list(day)
y=list(high)
z=list(low)
plt.scatter(y,z,color='crimson')
plt.plot(y,z,color='purple')
plt.xlabel("High")
plt.ylabel("Low")
plt.title('Bitcoin 10 days Graph')
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot(x,y,z,color='limegreen')
ax.scatter(x,y,z,color='royalblue',linewidths=2)
ax.set_xlabel('Day',color='red')
ax.set_ylabel('Price',color='indigo')
ax.set_zlabel('Profit/Loss',color='dodgerblue')
ax.set_facecolor('lightcyan')
ax.set_xlim(0,10.5)

ax.set_title('Bitcoin 10 days Graph')
plt.show()
```
